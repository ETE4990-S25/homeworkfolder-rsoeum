````python
rates = ["EUR", "GBP", "USD", "DZD", "AUD", "BWP", "BND", "CAD", "CLP", "CNY", "COP", "CZK", "DKK", "HUF", "ISK", "INR", "IDR", "ILS", "KZT", "KRW", "KWD", "LYD", "MYR", "MUR", "NPR", "NZD", "NOK", "OMR", "PKR", "PLN", "QAR", "RUB", "SAR", "SGD", "ZAR", "LKR", "SEK", "CHF", "THB", "TTD"]
ratesForBase = [r for r in rates if r != "USD" and r != "EUR" and r != "GBP"]
````

````python
import requests
import xmltodict
import json
import random

# URL of the XML data
date = "2011-05-04"
base = random.choice(ratesForBase)
url = f"https://www.floatrates.com/historical-exchange-rates.html?operation=rates&pb_id=1775&page=historical&currency_date={date}&base_currency_code={base}&format_type=xml"
print(url)
# Fetch the XML data
response = requests.get(url)
response.raise_for_status()  # Ensure we notice bad responses

# Parse the XML data to a Python dictionary
data_dict = xmltodict.parse(response.text)

# Convert the dictionary to a JSON string
json_data = json.dumps(data_dict, indent=4)

# Print the JSON data
print(json_data)

# Optionally, write the JSON data to a file
with open(f"{date}_exchange_rates_{base}.json", "w") as json_file:
    json_file.write(json_data)
````

````python
import os
import json
from datetime import datetime

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#had to hardcode to the directory i was having a lot of troubles loading the jsonfiles.
DATA_DIR = r"C:\Users\rvsoe\OneDrive\Desktop\ETE4990\homeworkfolder-rsoeum-1\final take home\data_random"

def find_item(obj):
    if isinstance(obj, dict):
        if "item" in obj:
            return obj["item"]
        for v in obj.values():
            res = find_item(v)
            if res is not None:
                return res
    elif isinstance(obj, list):
        for e in obj:
            res = find_item(e)
            if res is not None:
                return res
    return None

rows = []
for base in os.listdir(DATA_DIR):
    bd = os.path.join(DATA_DIR, base)
    if not os.path.isdir(bd):
        continue

    for fn in os.listdir(bd):
        if not fn.lower().endswith(".json"):
            continue
        ds = fn.split("_")[0]
        try:
            dt = datetime.strptime(ds, "%Y-%m-%d")
        except ValueError:
            continue
        doc = json.load(open(os.path.join(bd, fn), "r"))
        items = find_item(doc)
        if items is None:
            continue
        if isinstance(items, dict):
            items = [items]
        for ent in items:
            tgt = ent.get("@code") or ent.get("targetCurrency")
            val = ent.get("#text") or ent.get("exchangeRate")
            try:
                rate = float(val)
            except:
                continue
            rows.append({
                "date":     dt,
                "base":     base,
                "currency": tgt,
                "rate":     rate
            })

df = pd.DataFrame(rows)
print(f"loaded {len(df)} rows")
print("columns present:", df.columns.tolist())
print("sample rows:\n", df.head(), sep="")

#5 vs USD
chosen = ["BND","COP","DZD","HUF","KRW"]
df5 = df[(df["base"].isin(chosen)) & (df["currency"] == "USD")].copy()
ts = df5.pivot(index="date", columns="base", values="rate").sort_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=ts)
plt.title("Selected Bases → USD (2011–2023)")
plt.xlabel("Date"); plt.ylabel("USD per unit of base")
plt.legend(title="Base")
plt.tight_layout()
plt.show()

#yearly volatility
rets = ts.pct_change().abs()
vol_counts = (rets > 0.02).sum(axis=1).resample("Y").sum()
plt.figure(figsize=(10,5))
sns.barplot(x=vol_counts.index.year, y=vol_counts.values)
plt.title("Days per Year with >2% Moves")
plt.xlabel("Year"); plt.ylabel("Volatile Days")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#correlationheatmap
plt.figure(figsize=(8,6))
sns.heatmap(ts.corr(), annot=True, fmt=".2f", cbar_kws={"shrink":.8})
plt.title("correlation between BND/COP/DZD/HUF/KRW to USD")
plt.tight_layout()
plt.show()
````

Observations:
-In the line plot BND(Brunei Dollar) trades at 0.80 usd per BND. This is the highest out of the 5 being compared. 
-All the rates being analyzed are very flat which shows there were no spikes or crashes. 

-in the bar graph it shows that in the 5 selected currencies there were only 2 volatile days in 2011.

-in the heat map it shows that there is a strong correlation with all 5 currencies and the US. They generally move the same direction vs the US.

-HUF and DZD have the highest correlation at 0.90

-COP and KRW have the weakest correlation at 0.62
